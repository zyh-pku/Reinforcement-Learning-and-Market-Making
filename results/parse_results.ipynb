```python
def parse_file(file_path, value_threshold=20000):
    smaller_values = []
    with open(file_path, 'r') as file:
        for line in file:
            if "RL_Iteration_Steps:" in line:
                try:
                    value = int(line.split(": ")[1].strip())
                    if value < value_threshold:
                        smaller_values.append(value)
                except ValueError:
                    continue  # skip lines where conversion to int fails
    return smaller_values
```

```python
import numpy as np
# Example usage:
for i in range(10):
    file_path = f'hyperparameter_results_NP2_NY1_Delta{i}.txt'
    result = parse_file(file_path, value_threshold=20000)
    print(i, result)
```
